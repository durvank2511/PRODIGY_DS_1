```python
import pandas as pd
import matplotlib.pyplot as plt


data_path = '/kaggle/input/prodigy-ds-1/API_SP.POP.TOTL_DS2_en_csv_v2_1083038.csv'

data_df = pd.read_csv(data_path, skiprows=4)

data_2023 = data_df[['Country Name', '2023']].dropna()

data_2023 = data_2023.sort_values(by='2023', ascending=False).head(30)  # Select top 30 countries

plt.figure(figsize=(14, 8))
plt.bar(data_2023['Country Name'], data_2023['2023'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population in 2023')
plt.title('Population Distribution for Top 30 Countries in 2023')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```
